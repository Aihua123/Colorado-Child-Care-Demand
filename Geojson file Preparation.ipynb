```python
import requests
import pandas as pd
import json
import glob
from pprint import pprint
```

```python
URL = "https://raw.githubusercontent.com/OpenDataDE/State-zip-code-GeoJSON/master/co_colorado_zip_codes_geo.min.json"
r = requests.get(url = URL) 
data = r.json() 
```

```python
result = data.get('features')
result[1]
```

```python
def update_geojson(file):
    with open('result_data/'+ file) as f:
        result = json.load(f)
        
    for x in result:
        for y in data.get('features'):
            if y.get('properties').get('ZCTA5CE10') == str(x.get('ZIP')):
                y.get('properties').update(x)
```

```python
files = glob.glob('result_data/*.json')
file_names = [x.split("\\")[1] for x in files]


for f in file_names:
    update_geojson(f)
```
